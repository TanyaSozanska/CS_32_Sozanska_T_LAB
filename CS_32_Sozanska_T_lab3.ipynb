```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
from scipy.stats import norm
from sklearn.model_selection import train_test_split
import plotly.express as px

inf = pd.read_csv('IrisData_full.csv', sep=',', encoding= 'utf-8')
inf = inf.sample(frac=1)
inf.to_csv('iris1.csv', index=False)
numeric_columns = inf.select_dtypes(include=[np.number])

inf[numeric_columns.columns] = (numeric_columns - numeric_columns.min()) / (numeric_columns.max() - numeric_columns.min())

inf.to_csv('normalized_iris.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import numpy as np

inf = pd.read_csv('IrisData_full.csv', sep=',', encoding= 'utf-8')

# Перемішати записи у датасеті
d = np.random.permutation(inf)==inf

# Розділити датасет на навчальний і тестовий набори

train_set, test_set = train_test_split(inf, test_size=0.2)

fig = px.scatter_3d(inf, x='5.1', y='3.5',z='1.4', color='Iris-setosa', size='0.2')
fig.update_layout(title='Повний набір даних')
fig.show()
fig = px.scatter_3d(train_set, x='5.1', y='3.5',z='1.4', color='Iris-setosa', size='0.2')
fig.update_layout(title='Навчальна вибірка')
fig.show()
fig = px.scatter_3d(test_set, x='5.1', y='3.5',z='1.4', color='Iris-setosa', size='0.2')
fig.update_layout(title='Тестова вибірка')
fig.show()
# Вибрати ознаки та цільову змінну для навчального та тестового наборів
train_X = train_set[['5.1', '3.5', '1.4', '0.2']]
train_y = train_set['Iris-setosa']

test_X = test_set[['5.1', '3.5', '1.4', '0.2']]
test_y = test_set['Iris-setosa']

# Масштабування ознак
scaler = StandardScaler()
train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)

for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(train_X, train_y)

    train_pred = knn.predict(train_X)
    test_pred = knn.predict(test_X)

    train_accuracy = accuracy_score(train_y, train_pred)
    test_accuracy = accuracy_score(test_y, test_pred)

    print("При K =", i)
    print("Точність навчальної вибірки =", train_accuracy)
    print("Точність тестової вибірки =", test_accuracy)
```
